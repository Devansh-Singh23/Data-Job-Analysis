<h3>How are in-demand skills trending for Data Analysts?

Methodology
1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

```python
# Importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

# Loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] =='Data Analyst')&(df['job_country']=='United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)

df_DA_US_pivot.loc['Total']=df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()

DA_totals
```

```python
df_DA_US_percent=df_DA_US_pivot.div(DA_totals/100,axis=0)

# changes month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5].copy()
df_plot['x'] = range(len(df_plot))  # numeric x values

# Set seaborn theme and plot
sns.set_theme(style='ticks')

# Plot each line manually with numeric x-axis
ax = plt.gca()
palette = sns.color_palette('tab10')

for i, skill in enumerate(df_plot.columns[:-1]):  # skip 'x' column
    ax.plot(df_plot['x'], df_plot[skill], label=skill, color=palette[i])

sns.despine()
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.legend().remove()

# Annotate the last point
last_x = df_plot['x'].iloc[-1]
for i, skill in enumerate(df_plot.columns[:-1]):
    y = df_plot[skill].iloc[-1]
    if skill.lower() == 'tableau':
        y += 1
    elif skill.lower() == 'python':
        y -= 1
    ax.text(last_x + 0.3, y, skill, color='black', fontsize=10, va='center')

plt.show()
```
