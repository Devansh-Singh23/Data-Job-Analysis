<h4>What are the most demanded skills for the top 3 most popular data roles?

Methodology
1. Clean-up skill column
2. Calculate skill based on job_title_short
3. Calculate skill percentage
4. Plot final findings

```python
# Importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[df['job_country']=='United States'].copy()
```

```python
df_skills = df_US.explode('job_skills')

df_skills[['job_title_short','job_skills']]
```

```python
df_skills_count = df_skills.groupby(['job_skills','job_title_short']).size()

df_skills_count = df_skills_count.reset_index(name='skill_count')

df_skills_count.sort_values(by='skill_count', ascending=False, inplace= True)

df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])

job_titles
```

```python
fig, ax = plt.subplots(len(job_titles),1)

for i, job_titles in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short']==job_titles].head(5)
    df_plot.plot(kind ='barh',x='job_skills',y='skill_count',ax=ax[i],title=job_titles)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('Count of top skills in Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5)
plt.show()
```

```python
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total')
```

```python
df_skills_perc = pd.merge(df_skills_count,df_job_title_count,how='left',on='job_title_short')

df_skills_perc['skills_perc']=100* df_skills_perc['skill_count']/ df_skills_perc['jobs_total']

df_skills_perc
```

```python
import seaborn as sns

fig, ax = plt.subplots(len(job_titles),1)

sns.set_theme(style='ticks')

for i, job_titles in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short']==job_titles].head(5)
    #df_plot.plot(kind ='barh',x='job_skills',y='skills_perc',ax=ax[i],title=job_titles)
    sns.barplot(data =df_plot, x='skills_perc',y='job_skills',ax=ax[i],hue='skill_count',palette='dark:b_r')
    ax[i].set_title(job_titles)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().set_visible(False)
    ax[i].set_xlim(0, 78)

    if i != len(job_titles) - 1:
        ax[i].set_xticks([])

    for n, v in enumerate(df_plot['skills_perc']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')


fig.suptitle('Likelihood of skills requested in US Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5)
plt.show()
```
